```
def ME(a, b, n):
  c = 0
  d = 1
  bVector = bin(b)[2:]
  for i in range(len(bVector)):
    c *= 2
    d = (d*d) % n
    if (bVector[i]=='1'):
      c += 1
      d = (d * a) % n
  return d

def ME2(a, b, n):
  c = 0
  d = 1
  bVector = bin(b)[:1:-1]
  k = len(bVector)-1
  print (bVector, bin(b))
  for i in range(k,-1,-1):
    c *= 2
    d = (d*d) % n
    if (bVector[i]=='1'):
      c += 1
      d = (d * a) % n
  return d

a=111; b=113; n=97
a=2; b=521; n=5217
print (a**b%n, ME(a, b, n), ME2(a, b, n))
```

```
### Witness

def Witness(a, n):
  #t = [i for i in range(int(np.log2(n-1)),-1,-1) if ((n-1)%(2**i))==0][0]
  t = [i for i in range(len(bin(n))-3, -1, -1) if ((n-1)%(2**i))==0][0]
  u = (n-1)//(2**t)
  ## print (n-1,u,t, u*2**t)
  x1 = ME(a, u, n)
  ## print (x1)
  for i in range(t):
    x2 = (x1**2) % n
    ## print (x1, x2)
    if x2 == 1 and x1 != 1 and x1 != n-1:
      return True
    x1 = x2
    ## print (x1)
  if x1 != 1:
    return True  
  return False

n = int(np.random.randint(1000,10000))
print (n,list(range(int(np.log2(n-1)),-1,-1) ))

print (Witness( 5, n ))
```

```
def MillerRabin(n, s):
  for j in range(s):
    #a = np.random.randint(1,n-1)
    a = round(np.random.random()*(n-2))+1
    if (Witness(a,n)==True):
      Qprint (n, "COMPOSITE:", a)
      return False
  Qprint (n, "PRIME")
  return True

QUIET = True
for i in range(1000):
  x = np.random.randint(1000000,10000000)
  if x%2==0:
    continue
  if (MillerRabin(x, 3)==True):
    Qprint(x,"PRIME")
    break
```

```
def FindPrime(bits):
  for i in range(10000):
    x = int( np.random.random() * ((2**(bits-1))-1) + (2**(bits-1)) )
    if x%2==0:
      continue
    if (MillerRabin(x, 3)==True):
      print(x,"PRIME")
      return x

QUIET = False 
Primes = [FindPrime(50) for _ in range(5)]
for p in Primes:
  print (p, len(bin(p))-2)
```
